##ONe Hot Encoding
Setiap Ketgori pada suatu fitur akan di urutkan secara alfabet dan direpresentasikan sebagai kumpulan bits.

```python
data_NBA=data_NBA.drop(['Unnamed: 0'],axis=1)
data_NBA.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
#pada OneHotEncoder() kita tampung ke dalam variable one_hot_encoder
#next pada fit_tansform(menskalakan data)
#kita lihat pada X yaitu nilai yg terdapat pada atribut Player
#lalu hasilnya kita konversikan dalam satu array
one_hot_encoder = OneHotEncoder()
X = one_hot_encoder.fit_transform(X).toarray()
X
```

```python
#kita membentuk object DataFrama, dimana objeknya dibentuk dalam sekumpulan nilai X
df_one_hot = pd.DataFrame(X, columns=[str(i)for i in range(X.shape[1])])
df_one_hot
#kita bisa lihat data frame dibawah ini ,dmna pada player satu memiliki representasi nilai dibawah ini
```

```python
#kita gabung data frame dia atas menggunakan pd.concat
#pd.concat yaitu melakukan operasi penggabungan bersama pada suatu objek
#pada parameter axis kita beri nilai 1 agar hasilnya saling berdampingan
data_NBA = pd.concat([df_one_hot,data_NBA],axis=1)
data_NBA.head(10)
#disini kita liat Player pada peringkat 460, kita lihat direpresentasikan pada baris pertama 
```

##Label Encoding vs One Hot Encoding
Kita menerapkan Label Encoding jika:
1.   Nilai Categorical adalah nominal
2.   Jumlah kategori yang ada tidak terlalu banyak

Kita menerapkan One Hot Encoding jika:
1.  Nilai Categorical adalah Nominal 
2.  Jumlah Kategori yang ada tidak terlalu banyak
